## 1- Get Products Dataframe

```python
# getting only needed columns
products_df = df[['id', 'name', 'asins']].copy()
# drop duplicate rows, considering 'name' to identify them
products_df = products_df.drop_duplicates(subset='name')

products_df.head()
```

## 2- Get Users Ratings DataFrame

```python
users_df = df[['name','reviews.username', 'reviews.rating']].copy()
users_df.rename(columns={"reviews.username":"username", "reviews.rating":"rating"}, inplace=True)
users_df.head()
```

```python
users_df.shape
```

## 3- Get User Input

```python
user_input = [
    {'name': 'All-New Fire HD 8 Tablet 8 HD Display Wi-Fi 16 GB - Includes Special Offers Magenta', 'rating': 4.5},
    {'name': 'Fire Tablet 7 Display Wi-Fi 8 GB - Includes Special Offers Magenta', 'rating': 3},
    {'name': 'Brand New Amazon Kindle Fire 16gb 7 Ips Display Tablet Wifi 16 Gb Blue', 'rating': 4},
    {'name': 'Kindle Paperwhite E-reader - White 6 High-Resolution Display (300 ppi) with Built-in Light Wi-Fi - Includes Special Offers', 'rating': 3.5},
    {'name': 'New Amazon Kindle Fire Hd 9w Powerfast Adapter Charger + Micro Usb Angle Cable', 'rating': 2}
]

input_products = pd.DataFrame(user_input)
input_products
```

```python
input_products = input_products.sort_values(by='name')
input_products
```

## 4- Similarity Weights (Pearson Correlation)

```python
users_subset = users_df[users_df['name'].isin(input_products['name'].tolist())]
users_subset.head()
```

```python
users_subset.shape
```

```python
users_subset_group = users_subset.groupby(['username'])
```

```python
users_subset_group = sorted(users_subset_group,  key=lambda x: len(x[1]), reverse=True)
```

```python
users_subset_group = users_subset_group[:150]
```

```python
pearson_correlation_dict = {}

for name, group in users_subset_group:
    group = group.drop_duplicates('name')
    group = group.sort_values(by='name')
    temp_df = input_products[input_products['name'].isin(group['name'].tolist())]
    temp_rating_list = temp_df['rating'].tolist()
    temp_group_list = group['rating'].tolist()
    
    pearson_correlation_dict[name] = round(pearsonr(temp_rating_list, temp_group_list)[0], 3)
    
```

```python
pearson_correlation_dict.items()
```

```python
pearson_df = pd.DataFrame.from_dict(pearson_correlation_dict, orient='index').reset_index(drop=False)
pearson_df.columns = ['username', 'similarityIndex']
pearson_df = pearson_df.sort_values(by='similarityIndex',ascending=False)
pearson_df.head()
```

```python
top_users = pearson_df.iloc[:20]
top_users.head()
```

## 5- Find Recommendations

```python
top_users_rating = pd.merge(top_users, users_df, left_on='username', right_on='username', how='inner')
top_users_rating.head()
```

```python
top_users_rating.shape 
```

```python
top_users_rating['weightedRating'] = top_users_rating['similarityIndex']*top_users_rating['rating']
top_users_rating.head()
```

```python
temp_users_rating = top_users_rating.groupby('name').sum()[['similarityIndex', 'weightedRating']]
temp_users_rating.columns = ['sum_similarityIndex', 'sum_weightedRating']
temp_users_rating.head()
```

```python
recommendation_df = pd.DataFrame()
recommendation_df['weighted average recommendation score'] = temp_users_rating['sum_weightedRating']/temp_users_rating['sum_similarityIndex']
recommendation_df['name'] = temp_users_rating.index
recommendation_df = recommendation_df.reset_index(drop=True)
recommendation_df
```

```python
recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
recommendation_df.head(20)
```
