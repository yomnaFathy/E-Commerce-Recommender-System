## 1- Get Products with Categories Dataframe

```python
# getting only needed columns
products_df = df[['id', 'name', 'asins', 'categories']]
# drop duplicate rows, considering 'name' to identify them
products_df = products_df.drop_duplicates(subset='name')

products_df.head()
```

```python
products_df.shape
```

```python
# cleaning product name
products_df['name'] = products_df['name'].str.replace('[\,\\r]', '')
products_df['name'] = products_df['name'].str.replace('[\\n]', ' ')
products_df['name'] = products_df['name'].str.strip()
products_df.head()
```

```python
products_df['categories'] = products_df['categories'].str.split(',')
products_df.head()
```

```python
productsWithCategories_df = products_df.copy()

for index, row in productsWithCategories_df.iterrows():
    for category in row['categories']:
         productsWithCategories_df.at[index, category] = 1
            
productsWithCategories_df.head()
```

```python
productsWithCategories_df = productsWithCategories_df.fillna(0)
productsWithCategories_df.head()
```

## 2- Get User Input

```python
userInput = [
    {'name': 'All-New Fire HD 8 Tablet 8 HD Display Wi-Fi 16 GB - Includes Special Offers Magenta', 'rating': 4.5},
    {'name': 'Fire Tablet 7 Display Wi-Fi 8 GB - Includes Special Offers Magenta', 'rating': 3},
    {'name': 'Brand New Amazon Kindle Fire 16gb 7 Ips Display Tablet Wifi 16 Gb Blue', 'rating': 4},
    {'name': 'Kindle Paperwhite E-reader - White 6 High-Resolution Display (300 ppi) with Built-in Light Wi-Fi - Includes Special Offers', 'rating': 3.5},
    {'name': 'New Amazon Kindle Fire Hd 9w Powerfast Adapter Charger + Micro Usb Angle Cable', 'rating': 2}
]

inputProducts = pd.DataFrame(userInput)
inputProducts
```

```python
products_df[products_df['name'].isin(inputProducts['name'])] 
```

```python
userProducts = productsWithCategories_df[productsWithCategories_df['name'].isin(inputProducts['name'])]
userProducts
```

## 3- Weighting Categories and Getting User Profile

```python
userProducts = userProducts.reset_index(drop=True)
```

```python
UserCategoryTable = userProducts.drop(['id', 'name', 'asins', 'categories'], axis=1)
UserCategoryTable
```

```python
inputProducts['rating']
```

```python
userProfile = UserCategoryTable.transpose().dot(inputProducts['rating'])
userProfile
```

## 4- Finding Recommendation 

```python
categoryTable = productsWithCategories_df.set_index(productsWithCategories_df['name'])
categoryTable = categoryTable.drop(['id', 'name', 'asins', 'categories'], axis=1)
categoryTable
```

```python
recommendationTable_df = ((categoryTable*userProfile).sum(axis=1))/(userProfile.sum())
```

```python
recommendationTable_df = (recommendationTable_df.reset_index()).rename(columns={0: 'weightedAvg'})
recommendationTable_df
```

```python
user_recommendationTable_df = pd.merge(products_df, recommendationTable_df)
user_recommendationTable_df
```

```python
user_recommendationTable_df = user_recommendationTable_df.sort_values('weightedAvg', ascending=False)
user_recommendationTable_df.head(20)
```
